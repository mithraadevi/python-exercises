```python
import random

#Get Target Host
print("=== Network Ping Simulator ===\n")
host = input("Enter hostname or IP address to ping: ").strip()

#Validate Input
if host == "":
    print("\nNo hostname entered. Please try again.\n")

else:
    #Print Ping Header
    print(f"\nPinging {host} with 4 packets...\n")

    TOTAL_PINGS    = 4
    SUCCESS_RATE   = 0.80   # 80% chance each ping succeeds
    attempts       = []     # stores response times for successful pings
    failed_pings   = 0

    #Simulate 4 Ping Attempts
    for i in range(1, TOTAL_PINGS + 1):

        #generate a random response time between 10ms and 200ms
        response_time = random.randint(10, 200)

        #randomly decide success or failure
        success = random.random() < SUCCESS_RATE   # random.random() gives a float between 0.0 and 1.0

        if success:
            print(f"  Reply from {host}: time={response_time}ms")
            attempts.append(response_time)
        else:
            print(f"  Request timeout for ping #{i}")
            failed_pings += 1

    #Calculate Statistics
    total_sent    = TOTAL_PINGS
    total_lost    = failed_pings
    total_success = len(attempts)
    loss_percent  = (total_lost / total_sent) * 100

    if total_success > 0:
        min_time = min(attempts)
        max_time = max(attempts)
        avg_time = sum(attempts) // len(attempts)   # integer division for clean ms display
    else:
        min_time = max_time = avg_time = 0

    #Display Results
    print(f"\n--- Ping statistics for {host} ---")
    print(f"  Packets: Sent = {total_sent}, Received = {total_success}, Lost = {total_lost} ({loss_percent:.0f}% loss)")

    if total_success > 0:
        print(f"\n  Approximate round trip times in milliseconds:")
        print(f"  Minimum = {min_time}ms | Maximum = {max_time}ms | Average = {avg_time}ms")

    #Status Message
    print()
    if total_success == 0:
        print("Status: Host is unreachable. All packets lost.")
    elif loss_percent > 25 or avg_time > 100:
        print("Status: Host is reachable but experiencing network issues.")
    else:
        print("Status: Host is highly reachable. Connection looks stable.")
    print()
```
