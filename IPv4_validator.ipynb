```python
def is_valid_ip(ip):
    parts = ip.split(".")

    #Must have exactly 4 parts
    if len(parts) != 4:
        return False

    for part in parts:
        #Empty part check (handles cases like "192..1.1")
        if part == "":
            return False

        if not part.isdigit():
            return False

        if int(part) < 0 or int(part) > 255:
            return False

    return True


#Main Program
print("=== IP Address Validator ===")
ip = input("Enter an IP address: ").strip()

if is_valid_ip(ip):
    print(f"\n{ip} is a valid IPv4 address.\n")
else:
    print(f"\n{ip} is NOT a valid IPv4 address.\n")
```
