```python
import re

# List of common passwords to check against
COMMON_PASSWORDS = [
    "1234", "12345", "123456", "1234567", "12345678",
    "password", "password123", "pass123", "qwerty", "qwerty123",
    "abc123", "admin", "admin123", "letmein", "welcome",
    "iloveyou", "monkey", "dragon", "master", "hello",
    "sunshine", "princess", "shadow", "superman", "batman"
]

def check_password_strength(password):
    # --- Step 1: Check if it's a common password ---
    if password.lower() in COMMON_PASSWORDS:
        print("\nâš ď¸Ź  This is a very common password!")
        print("Please try a stronger and more unique password.\n")
        return

    # --- Step 2: Run strength checks ---
    issues = []

    # Length check (minimum 8 characters recommended)
    if len(password) < 8:
        issues.append("make it at least 8 characters long")

    # Uppercase letter check
    if not re.search(r'[A-Z]', password):
        issues.append("add at least one uppercase letter (A-Z)")

    # Lowercase letter check
    if not re.search(r'[a-z]', password):
        issues.append("add at least one lowercase letter (a-z)")

    # Number check
    if not re.search(r'[0-9]', password):
        issues.append("include at least one number (0-9)")

    # Special character check
    if not re.search(r'[!@#$%^&*()_+\-=\[\]{};\':",.<>?/\\|`~]', password):
        issues.append("use at least one special character (e.g. @, #, $, !)")

    # --- Step 3: Give feedback based on results ---
    if len(issues) == 0:
        print("\nâś… Your password looks strong! Great job.\n")
    else:
        print("\nâšˇ Your password is moderate. You can improve it by:")
        for tip in issues:
            print(f"   â†’ {tip}")
        print()

# --- Main Program ---
print("=== Password Strength Checker ===")
password = input("Enter your password: ")
check_password_strength(password)
```
