```python
import hashlib

def get_hash(data_bytes, algorithm):
    # Compute hash based on chosen algorithm
    if algorithm == "md5":
        return hashlib.md5(data_bytes).hexdigest()
    elif algorithm == "sha1":
        return hashlib.sha1(data_bytes).hexdigest()
    elif algorithm == "sha256":
        return hashlib.sha256(data_bytes).hexdigest()
    else:
        return None


def hash_text(text, algorithm):
    data_bytes = text.encode('utf-8')  # convert text to bytes before hashing
    return get_hash(data_bytes, algorithm)


def hash_file(filepath, algorithm):
    with open(filepath, 'rb') as f:
        data_bytes = f.read()  # file is read directly as bytes
    return get_hash(data_bytes, algorithm)


def compare_hashes(hash1, hash2):
    # Plain string comparison -> case-sensitive by default
    return hash1 == hash2


# Main Program
print("=== Hash Checker ===")
print("Choose hash algorithm:")
print("1. MD5")
print("2. SHA1")
print("3. SHA256")

algo_choice = input("Enter choice (1/2/3): ").strip()
algo_map = {"1": "md5", "2": "sha1", "3": "sha256"}
algorithm = algo_map.get(algo_choice)

if not algorithm:
    print("\nInvalid choice. Please run again and choose 1, 2, or 3.\n")
else:
    print("\nWhat do you want to hash?")
    print("1. Text input")
    print("2. File")
    input_choice = input("Enter choice (1/2): ").strip()

    computed_hash = None

    if input_choice == "1":
        text = input("Enter the text: ")
        computed_hash = hash_text(text, algorithm)

    elif input_choice == "2":
        filepath = input("Enter the file path: ").strip()
        try:
            computed_hash = hash_file(filepath, algorithm)
        except FileNotFoundError:
            print("\nFile not found. Please check the path and try again.\n")

    else:
        print("\nInvalid choice.\n")

    if computed_hash:
        print(f"\nComputed {algorithm.upper()} hash: {computed_hash}")

        compare_choice = input("\nDo you want to compare this with another hash value? (y/n): ").strip().lower()
        if compare_choice == "y":
            other_hash = input("Enter the hash value to compare: ").strip()
            if compare_hashes(computed_hash, other_hash):
                print("\nThe hashes match!\n")
            else:
                print("\nThe hashes do NOT match.\n")
```
