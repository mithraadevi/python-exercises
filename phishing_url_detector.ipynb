```python
# ===== Phishing URL Detector =====

# Setup
risk_score = 0
flags = []

print("=== Phishing URL Detector ===\n")
url = input("Enter a URL to scan: ").strip().lower()

# --- Protocol Check ---
if not url.startswith("https://"):
    risk_score += 1
    flags.append("Insecure protocol — URL does not use HTTPS")

# --- Suspicious Keyword Scan ---
suspicious_keywords = [
    "login", "verify", "secure", "account", "update",
    "confirm", "paypal", "password", "signin", "banking",
    "ebay", "amazon", "support", "free", "lucky", "winner"
]

for keyword in suspicious_keywords:
    if keyword in url:
        risk_score += 1
        flags.append(f"Suspicious keyword found: '{keyword}'")
        break

# --- IP Address Detection ---
domain = url.replace("https://", "").replace("http://", "")
domain = domain.split("/")[0]

parts = domain.split(".")
is_ip = False

if len(parts) == 4:
    is_ip = True
    for part in parts:
        if not part.isdigit() or int(part) > 255:
            is_ip = False
            break

if is_ip:
    risk_score += 3
    flags.append("CRITICAL — URL uses a raw IP address instead of a domain name")

# --- URL Length Check ---
if len(url) > 75:
    risk_score += 1
    flags.append("Unusually long URL — possible obfuscation or multiple redirects")

# --- Print Security Report ---
print("\n" + "="*40)
print("     SECURITY REPORT")
print("="*40)
print(f"URL Scanned : {url}")
print(f"Risk Score  : {risk_score}")

if risk_score == 0:
    verdict = "SAFE — No suspicious patterns detected"
elif risk_score == 1:
    verdict = "CAUTION — Minor concerns detected. Verify before clicking"
elif risk_score <= 3:
    verdict = "HIGH RISK — Multiple red flags. Avoid this link"
else:
    verdict = "CRITICAL — Strong phishing indicators. Do NOT click"

print(f"Verdict     : {verdict}")

if flags:
    print("\nFlags Triggered:")
    for flag in flags:
        print(f"   → {flag}")
else:
    print("\nNo suspicious flags detected.")

print("="*40 + "\n");
```
