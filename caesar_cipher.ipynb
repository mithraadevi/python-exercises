```python
def caesar_cipher(text, key, mode):
    # If decoding, reverse the shift direction
    if mode == "decode":
        key = -key

    result = ""

    for char in text:
        if char.isupper():
            # Shift within A-Z (0-25), wrap using modulo
            shifted = (ord(char) - ord('A') + key) % 26
            result += chr(shifted + ord('A'))

        elif char.islower():
            # Shift within a-z (0-25), wrap using modulo
            shifted = (ord(char) - ord('a') + key) % 26
            result += chr(shifted + ord('a'))

        else:
            # Leave numbers, spaces, symbols untouched
            result += char

    return result


# --- Main Program ---
print("=== Caesar Cipher ===")
choice = input("Do you want to (E)ncode or (D)ecode? ").strip().lower()
text = input("Enter your text: ")
key = int(input("Enter the key (shift number): "))

if choice == "e":
    output = caesar_cipher(text, key, "encode")
    print(f"\nEncoded text: {output}\n")
elif choice == "d":
    output = caesar_cipher(text, key, "decode")
    print(f"\nDecoded text: {output}\n")
else:
    print("\nInvalid choice. Please enter 'E' or 'D'.\n")
```
