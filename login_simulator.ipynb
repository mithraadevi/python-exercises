```python
# Hardcoded credentials (in real systems these would be stored securely)
CORRECT_USERNAME = "admin"
CORRECT_PASSWORD = "secure123"

MAX_ATTEMPTS = 3
attempts = 0

print("=== Login System ===\n")

while attempts < MAX_ATTEMPTS:
    username = input("Enter username: ")
    password = input("Enter password: ")

    # Both username AND password must match
    if username == CORRECT_USERNAME and password == CORRECT_PASSWORD:
        print(f"\n Login successful! Welcome, {username}.\n")
        break

    else:
        attempts += 1
        remaining = MAX_ATTEMPTS - attempts

        if remaining == 0:
            print("\nAccount locked! Too many failed attempts.")
            print("Please try again in 30 seconds.\n")
        else:
            print(f"\nWrong credentials. You have {remaining} attempt(s) remaining.\n")
```
