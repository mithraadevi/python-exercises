```python
import base64

def xor_encrypt_decrypt(message, key):
    result = ""
    for i in range(len(message)):
        msg_char  = ord(message[i])         # ASCII value of message character
        key_char  = ord(key[i % len(key)])  # ASCII value of key character (repeats using modulo)
        xor_value = msg_char ^ key_char     # XOR the two values
        result   += chr(xor_value)          # convert result back to character
    return result


def encrypt_message(message, key):
    #XOR encrypt the message
    encrypted = xor_encrypt_decrypt(message, key)

    #Convert to base64 so it displays as readable text instead of gibberish
    #base64 converts binary data into a text-safe format using A-Z, a-z, 0-9, +, /, =
    encoded = base64.b64encode(encrypted.encode('latin-1')).decode('utf-8')

    return encoded


def decrypt_message(encoded_message, key):
    #Convert from base64 back to original encrypted form
    encrypted = base64.b64decode(encoded_message).decode('latin-1')

    #XOR decrypt using the same key
    decrypted = xor_encrypt_decrypt(encrypted, key)

    return decrypted


#Main Program
print("=== XOR Encryption / Decryption Tool ===\n")
print("How it works:")
print("  XOR uses a secret key to scramble your text into ciphertext.")
print("  The exact same key unscrambles it back to plaintext.")
print("  This is called Symmetric Encryption — one key does both jobs.\n")

#Get User Choice
choice = input("Do you want to (E)ncrypt or (D)ecrypt? ").strip().lower()

if choice not in ["e", "d"]:
    print("\nInvalid choice. Please enter E or D.\n")

else:
    #Get the Message
    if choice == "e":
        message = input("Enter your message to encrypt: ")
    else:
        message = input("Enter your encrypted message (base64 format): ")

    #Get the Secret Key
    key = input("Enter your secret key: ")

    #Validate Inputs
    if message == "":
        print("\nMessage cannot be empty. Please try again.\n")
    elif key == "":
        print("\nKey cannot be empty. Please try again.\n")

    else:
        #Show Key Repetition
        if len(key) < len(message):
            repeated_key = ""
            for i in range(len(message)):
                repeated_key += key[i % len(key)]
            print(f"\nKey is shorter than message — repeating key: '{repeated_key}'")

        #Encrypt or Decrypt
        print("\n==============================")

        if choice == "e":
            result = encrypt_message(message, key)
            print("         Encryption Result      ")
            print("==============================")
            print(f"Original Message   : {message}")
            print(f"Secret Key Used    : {key}")
            print(f"Encrypted Output   : {result}")
            print("==============================")
            print("\nNote: The encrypted output is in base64 format.")
            print("This makes it readable and safe to copy/paste.")
            print("To decrypt, use this exact encrypted text and key.")

        else:
            result = decrypt_message(message, key)
            print("         Decryption Result      ")
            print("==============================")
            print(f"Encrypted Message  : {message}")
            print(f"Secret Key Used    : {key}")
            print(f"Decrypted Output   : {result}")
            print("==============================")

        print()
        print("Remember: XOR is reversible.")
        print("Encrypt your message, then decrypt it with the")
        print("same key — you'll get your original message back.")
        print()
```
