```python
import hashlib

print("=== File Integrity Checker ===")
print("Paste or type two pieces of text to check if they are identical.\n")

#Get Both Text Inputs
text1 = input("Enter first text  (File A): ")
text2 = input("Enter second text (File B): ")

#Validate Both Inputs Exist
if text1 == "" or text2 == "":
    print("\nError: Both texts must be provided. Please try again.\n")

else:
    #Convert Text to Bytes
    bytes1 = text1.encode('utf-8')
    bytes2 = text2.encode('utf-8')

    #Hash Both Texts Using SHA256
    hash1 = hashlib.sha256(bytes1).hexdigest()
    hash2 = hashlib.sha256(bytes2).hexdigest()

    #Display Both Hashes
    print("\n" + "=" * 50)
    print("                  HASH RESULTS")
    print("=" * 50)
    print(f"  File A Hash : {hash1}")
    print(f"  File B Hash : {hash2}")
    print("-" * 50)

    #Optional — show length of both texts
    print(f"  File A Length : {len(text1)} character(s)")
    print(f"  File B Length : {len(text2)} character(s)")
    print("-" * 50)

    #Compare the Hashes
    if hash1 == hash2:
        print("\n   MATCH — Both texts are identical.")
        print("     The hashes are exactly the same.")
    else:
        print("\n   MISMATCH — The texts are different.")
        print("     Even a single character change produces a completely")
        print("     different hash. That is what you are seeing here.")

    #Real World Explanation
    print()
    print("=" * 50)
    print("  How this works in real life:")
    print("  When you download a file, the website often publishes")
    print("  the official SHA256 hash of that file. You can hash")
    print("  your downloaded copy and compare — if the hashes")
    print("  match, the file is genuine and untampered.")
    print("  If they don't match, the file may be corrupted")
    print("  or replaced with a malicious version.")
    print("=" * 50)
    print()
```
